```python
df_clean = df.copy()
```

“Criei uma cópia do dataset para preservar os dados originais.”

```python
price_cols = ["actual_price", "discounted_price"]

for col in price_cols:
    df_clean[col] = (
        df_clean[col]
        .str.replace("₹", "", regex=False)
        .str.replace(",", "", regex=False)
        .astype(float)
    )
```

```python
df_clean[price_cols].info()
```

“Removi símbolos e converti preços para tipo numérico.”

```python
df_clean["discount_percentage"] = (
    df_clean["discount_percentage"]
    .astype(str)
    .str.replace("%", "", regex=False)
)

df_clean["discount_percentage"] = pd.to_numeric(
    df_clean["discount_percentage"], errors="coerce"
)

```

“Transformei percentual de desconto em variável numérica para análise.”

```python
df_clean["rating"] = pd.to_numeric(df_clean["rating"], errors="coerce")
df_clean["rating_count"] = (
    df_clean["rating_count"]
    .str.replace(",", "", regex=False)
    .astype(float)
)
```

“Usei coerção para lidar com valores inválidos de rating.”

```python
df_clean.isnull().sum()
```

```python
df_clean = df_clean.dropna(subset=["rating"])
df_clean["rating_count"] = df_clean["rating_count"].fillna(0)
```

“Removi registros sem avaliação e tratei volume ausente como zero.”

```python
df_clean["main_category"] = df_clean["category"].str.split("|").str[0]
```

“Criei uma variável de categoria principal para facilitar análise agregada.”

```python
df_clean["weighted_rating"] = (
    df_clean["rating"] * np.log1p(df_clean["rating_count"])
)
```

“Criei uma métrica de valor percebido ponderando avaliação pelo volume de avaliações.”

“Avaliei a distribuição de produtos para entender representatividade por categoria.”

```python
price_by_category = (
    df_clean
    .groupby("main_category")["discounted_price"]
    .mean()
    .sort_values(ascending=False)
)

price_by_category.head(10)
```

"Visualização para sequência da analise"

```python
import matplotlib.pyplot as plt

price_by_category.head(10).plot(kind="barh")
plt.title("Preço médio por categoria")
plt.xlabel("Preço médio (com desconto)")
plt.ylabel("Categoria")
plt.show()
```

Preço médio por categoria:

- Electronics concentra disparadamente os produtos mais caros.

- Car & Motorbike e Home & Kitchen aparecem em um segundo nível de preço.

- Categorias como Toys & Games, Office Products e Home Improvement operam com preços bem mais baixos.

Observação: O portfólio de produtos é altamente heterogêneo em preço, o que exige estratégias de pricing diferentes por categoria.

```python
discount_by_category = (
    df_clean
    .groupby("main_category")["discount_percentage"]
    .mean()
    .sort_values(ascending=False)
)

discount_by_category.head(10)
```

```python
discount_by_category.head(10).plot(kind="barh")
plt.title("Desconto médio por categoria")
plt.xlabel("Desconto médio (%)")
plt.ylabel("Categoria")
plt.show()
```

Desconto médio por categoria:

- Categorias com preços mais altos também apresentam descontos médios elevados.

- Isso sugere uso de desconto como estratégia para viabilizar produtos de maior ticket.

- Algumas categorias mais baratas aplicam descontos menores, preservando margem.

Observação: Desconto parece estar sendo usado como alavanca comercial, não como estratégia uniforme.

```python
plt.scatter(
    df_clean["discount_percentage"],
    df_clean["rating"],
    alpha=0.4
)

plt.xlabel("Desconto (%)")
plt.ylabel("Rating")
plt.title("Desconto vs Avaliação do Produto")
plt.show()
```

Desconto vs Avaliação:

- O gráfico mostra alta dispersão.

- Produtos com pouco desconto podem ter ratings altos.

- Produtos com desconto extremo nem sempre são melhor avaliados.

Observação: Não há relação forte entre aumento de desconto e melhoria da avaliação do cliente.

```python
plt.scatter(
    df_clean["discounted_price"],
    df_clean["rating"],
    alpha=0.4
)

plt.xlabel("Preço com desconto")
plt.ylabel("Rating")
plt.title("Preço vs Avaliação do Produto")
plt.show()
```

Preço vs Avaliação:

- Ratings altos aparecem em todas as faixas de preço.

- Produtos caros não são automaticamente melhor avaliados.

- Produtos baratos também conseguem excelente percepção.

Observação: Preço não é determinante direto de satisfação percebida.

```python
value_by_category = (
    df_clean
    .groupby("main_category")["weighted_rating"]
    .mean()
    .sort_values(ascending=False)
)

value_by_category.head(10)
```

```python
value_by_category.head(10).plot(kind="barh")
plt.title("Valor percebido médio por categoria")
plt.xlabel("Rating ponderado")
plt.ylabel("Categoria")
plt.show()
```

Valor percebido médio por categoria (rating ponderado):

- Categorias como Toys & Games e Musical Instruments se destacam em valor percebido.

- Essas categorias não são as mais caras nem as mais descontadas.

- Categorias com descontos agressivos nem sempre lideram em valor percebido.

Observação: Existe eficiência de pricing em algumas categorias, onde bom valor é entregue sem dependência de grandes descontos.

## 📌 Síntese dos Resultados


- A análise evidenciou que o portfólio de produtos apresenta grande heterogeneidade de preços e políticas de desconto entre categorias. Categorias de maior ticket médio, como Electronics, utilizam descontos elevados como estratégia comercial, enquanto categorias de menor preço operam com políticas mais conservadoras.

- Não foi identificada uma relação forte entre o percentual de desconto e a avaliação dos produtos, indicando que descontos elevados não garantem melhor percepção do cliente. De forma semelhante, produtos mais caros não apresentaram, em média, avaliações superiores.

- Por outro lado, a análise de valor percebido, utilizando rating ponderado pelo volume de avaliações, mostrou que algumas categorias entregam alta percepção de valor sem depender de descontos agressivos, sugerindo oportunidades de otimização de margem e revisão de estratégias de pricing.

- O principal insight observado foi que desconto e preço, isoladamente, não explicam a percepção do cliente. Algumas categorias conseguem entregar alto valor percebido sem depender de grandes descontos, o que indica espaço para otimização de margem e estratégias de pricing mais segmentadas.
